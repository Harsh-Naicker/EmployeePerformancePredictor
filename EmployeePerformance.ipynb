```
from sklearn.preprocessing import LabelEncoder
label=LabelEncoder()
label.fit(performance['Attrition'])
performance['Attrition']=label.transform(performance['Attrition'])
label.fit(performance['Travel'])
performance['Travel']=label.transform(performance['Travel'])
label.fit(performance['Dept'])
performance['Dept']=label.transform(performance['Dept'])
label.fit(performance['EducationField'])
performance['EducationField']=label.transform(performance['EducationField'])
label.fit(performance['Gender'])
performance['Gender']=label.transform(performance['Gender'])
label.fit(performance['Role'])
performance['Role']=label.transform(performance['Role'])
label.fit(performance['Marital'])
performance['Marital']=label.transform(performance['Marital'])
label.fit(performance['OverTime'])
performance['OverTime']=label.transform(performance['OverTime'])
label.fit(performance['Over18'])
performance['Over18']=label.transform(performance['Over18'])
```

```
performance.head()
```

```
X = performance.iloc[:, :-1].values
y=performance.iloc[:,34].values
```

```
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.30)
```

```
from sklearn.ensemble import RandomForestClassifier
classifier = RandomForestClassifier(n_estimators = 7)
classifier.fit(X_train, y_train)
```

```
y_pred = classifier.predict(X_test)
print(y_pred)
```

```
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
result = confusion_matrix(y_test, y_pred)
print("Confusion Matrix:")
print(result)
result1 = classification_report(y_test, y_pred)
print("Classification Report:",)
print (result1)
result2 = accuracy_score(y_test,y_pred)
print("Accuracy:",result2)
```
